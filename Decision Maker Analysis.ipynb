```python
import pandas as pd
```

```python
df=pd.read_csv("data_decision_maker.csv")
df=df.dropna()
df.head()
```

```python
df['Designation']=df['Designation'].astype(str).str.lower()
df["Designation Label"]=""
heads={'bureau':'head of bureau','tourism':'head of tourism department','communication':"head of customer service department",'sales':'head of sales department','marketing':'head of marketing department','planner':'head of event planner','wedding':'head of event planner','pr':"head of customer service department",'mice':'head of tourism department','default':'head of operations'}
director={'managing':'managing director','sales':'sales director','marketing':'marketing director','planner':'director of events and meeting','events':'director of events and meeting','regional':'director of regional branch','default':'director' }
owner={"co-owner":"co-owner","co":"co-owner","co-":"co-owner","default":'owner'}
president={"president":"president","vice":"vice-president","default":"president"}
founder={"co-founder":"co-founder","co":"co-founder","co-":"co-founder","default":'founder'}
manager={'general':'general manager','sales':'sales manager','marketing':'marketing manager','default':'manager' }
s_and_m={'sales':"sales and marketing representative",'marketing':"sales and marketing representative",'executive':"sales and marketing executive",'default':"sales and marketing representative"}
executive={'sales':"sales and marketing executive",'marketing':"sales and marketing executive",'default':"executive"}
for itr in range(0,len(df['Designation'])):
    if df['Designation'].iloc[itr].find('owner')!=-1:
        #print(df['Designation'].iloc[itr])
        designation=df['Designation'].iloc[itr].split(" ")
        #print(designation)
        flag=0
        for word in designation:
            if word in owner:
                df["Designation Label"].iloc[itr]=owner[word]
                flag=1
                break
        if flag==0:
            df["Designation Label"].iloc[itr]=owner['default']
    
    if df['Designation Label'].iloc[itr]=="":
        if df['Designation'].iloc[itr].find('founder')!=-1:
            #print(df['Designation'].iloc[itr])
            designation=df['Designation'].iloc[itr].split(" ")
            #print(designation)
            flag=0
            for word in designation:
                if word in founder:
                    df["Designation Label"].iloc[itr]=founder[word]
                    flag=1
                    break
            if flag==0:
                df["Designation Label"].iloc[itr]=founder['default']

    
    
    
    
    if df['Designation Label'].iloc[itr]=="":
        if (df['Designation'].iloc[itr].find('president')!=-1 or df['Designation'].iloc[itr].find('vp')!=-1 or df['Designation'].iloc[itr].find('svp')!=-1):
            #print(df['Designation'].iloc[itr])
            designation=df['Designation'].iloc[itr].split(" ")
            #print(designation)
            flag=0
            for word in designation:
                if word in president:
                    df["Designation Label"].iloc[itr]=president[word]
                    flag=1
                    break
            if flag==0:
                df["Designation Label"].iloc[itr]=president['default']

            
            
            
    if df['Designation Label'].iloc[itr]=="":
        if (df['Designation'].iloc[itr].find('director')!=-1 or df['Designation'].iloc[itr].find('directeur')!=-1 or df['Designation'].iloc[itr].find('direccio')!=-1) :
            #print(df['Designation'].iloc[itr])
            designation=df['Designation'].iloc[itr].split(" ")
            #print(designation)
            flag=0
            for word in designation:
                if word in director:
                    df["Designation Label"].iloc[itr]=director[word]
                    flag=1
                    break
            if flag==0:
                df["Designation Label"].iloc[itr]=director['default']
    if df['Designation Label'].iloc[itr]=="":
        if df['Designation'].iloc[itr].find('head')!=-1:
            #print(df['Designation'].iloc[itr])
            designation=df['Designation'].iloc[itr].split(" ")
            #print(designation)
            flag=0
            for word in designation:
                if word in heads:
                    df["Designation Label"].iloc[itr]=heads[word]
                    flag=1
                    break
            if flag==0:
                df["Designation Label"].iloc[itr]=heads['default']
    
    if df['Designation Label'].iloc[itr]=="":
        if df['Designation'].iloc[itr].find('ceo')!=-1:
            df["Designation Label"].iloc[itr]="ceo"
    
    if df['Designation Label'].iloc[itr]=="":
        if (df['Designation'].iloc[itr].find('manager')!=-1 or df['Designation'].iloc[itr].find('management')!=-1 or df['Designation'].iloc[itr].find('managr')!=-1 or df['Designation'].iloc[itr].find('man.')!=-1):
            #print(df['Designation'].iloc[itr])
            designation=df['Designation'].iloc[itr].split(" ")
            #print(designation)
            flag=0
            for word in designation:
                if word in manager:
                    df["Designation Label"].iloc[itr]=manager[word]
                    flag=1
                    break
            if flag==0:
                df["Designation Label"].iloc[itr]=manager['default']
    if df['Designation Label'].iloc[itr]=="":
        if (df['Designation'].iloc[itr].find('executive')!=-1 ):
            #print(df['Designation'].iloc[itr])
            designation=df['Designation'].iloc[itr].split(" ")
            #print(designation)
            flag=0
            for word in designation:
                if word in executive:
                    df["Designation Label"].iloc[itr]=executive[word]
                    flag=1
                    break
            if flag==0:
                df["Designation Label"].iloc[itr]=executive['default']
    
    if df['Designation Label'].iloc[itr]=="":
        if (df['Designation'].iloc[itr].find('planner')!=-1 and df['Designation'].iloc[itr].find('wedding')!=-1):
            df["Designation Label"].iloc[itr]="wedding planner"
    
                
    if df['Designation Label'].iloc[itr]=="":
        if (df['Designation'].iloc[itr].find('sales')!=-1 or df['Designation'].iloc[itr].find('marketing')!=-1 or df['Designation'].iloc[itr].find('executive')!=-1 ):
            #print(df['Designation'].iloc[itr])
            designation=df['Designation'].iloc[itr].split(" ")
            #print(designation)
            flag=0
            for word in designation:
                if word in s_and_m:
                    df["Designation Label"].iloc[itr]=s_and_m[word]
                    flag=1
                    break
            if flag==0:
                df["Designation Label"].iloc[itr]=s_and_m['default']
        
```
