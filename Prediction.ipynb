```python
import pandas as pd
import pickle
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")
```

```python
def label_encode(data,attribute):
    country=LabelEncoder() #data extracted from linkedin consist of only the following country
    country.fit(["High Income" , "Upper Middle Income" , "Lower Income"])
    
    size_category=LabelEncoder()#labels of size
    size_category.fit(['large enterprise', 'small and medium-sized enterprises'])
    
    category=LabelEncoder()#labels of category
    category.fit(['CB', 'CC', 'CH', 'EP', 'H', 'R', 'TA', 'TB', 'TO', 'WP'])
    
    
    if attribute=="country":
        return country.transform(data["country_category"])
    elif attribute=="size":
        return size_category
    elif attribute=="category":
        return category.transform(data["Category"])
    else:
        print("Invalid argument")
        
def prediction(dataframe):
    dataframe_final=dataframe.copy()
    dataframe["country_encoded"]=label_encode(dataframe,"country")
    dataframe["category_encoded"]=label_encode(dataframe,"category")
    size_encoder=label_encode(dataframe,"size")
    df_model=dataframe[["Rank","country_encoded","category_encoded"]]
    dump_file = 'automl.dump.pkl'
    model = pickle.load(open(dump_file, 'rb'))
    predictions=model.predict(df_model)
    dataframe_final["Predicted Size Category"]=size_encoder.inverse_transform(predictions)
    return dataframe_final
```

```python
df=pd.read_csv("predict.csv")
```
